```python
import pandas as pd
dfs = pd.read_excel(r"http://github.com/data-to-insight/ERN-sessions/raw/main/data/903_xlsx_errors.xlsx", sheet_name=None)
oc2 =dfs["oc2"]
oc2['DOB'] = pd.to_datetime(oc2["DOB"], format= "%d/%m/%Y", errors='coerce')
collection_End = pd.to_datetime('21/03/2020', dayfirst= True)
oc2['17bday'] =oc2['DOB'] + pd.DateOffset(years=17)
error_mask =(oc2['17bday']> collection_End) & (oc2['SDQ_SCORE'].notna()|oc2['SDQ_REASON'].notna())
errorrows=oc2[error_mask]
errorrows
```
